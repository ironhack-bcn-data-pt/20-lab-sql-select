## Challenge 1 - Who Have Published What At Where?

```python
%%sql
select a.au_id AUTHOR_ID, a.au_lname LAST_NAME, a.au_fname FIRST_NAME, t.title TITLE, p.pub_name PUBLISHER
from 
    authors a
    inner join 
        (titleauthor ta   
        inner join 
        (titles t inner join publishers p on t.pub_id = p.pub_id)
        on ta.title_id = t.title_id)
    on a.au_id = ta.au_id
;
```

## Challenge 2 - Who Have Published How Many At Where?

```python
%%sql
select a.au_id AUTHOR_ID, a.au_lname LAST_NAME, a.au_fname FIRST_NAME, p.pub_name PUBLISHER, count(t.title) "TITLE COUNT"
from 
    authors a
    inner join 
        (titleauthor ta   
        inner join 
        (titles t inner join publishers p on t.pub_id = p.pub_id)
        on ta.title_id = t.title_id)
    on a.au_id = ta.au_id
group by AUTHOR_ID
order by AUTHOR_ID DESC
;
```

## Challenge 3 - Best Selling Authors

```python
%%sql
select a.au_id AUTHOR_ID, a.au_lname LAST_NAME, a.au_fname FIRST_NAME, sum(t.ytd_sales) TOTAL
from 
    authors a
    left join 
        (titleauthor ta   
        inner join 
        (titles t inner join publishers p on t.pub_id = p.pub_id)
        on ta.title_id = t.title_id)
    on a.au_id = ta.au_id
group by AUTHOR_ID
order by TOTAL DESC
limit 3
;
```

## Challenge 4 - Best Selling Authors Ranking

```python
%%sql
select a.au_id AUTHOR_ID, a.au_lname LAST_NAME, a.au_fname FIRST_NAME, ifnull(sum(t.ytd_sales),0) TOTAL
from 
    authors a
    left join 
        (titleauthor ta   
        inner join 
        (titles t inner join publishers p on t.pub_id = p.pub_id)
        on ta.title_id = t.title_id)
    on a.au_id = ta.au_id
group by AUTHOR_ID
order by TOTAL DESC
;
```

## Bonus Challenge - Most Profiting Authors

```python
%%sql
select a.au_id AUTHOR_ID, a.au_lname LAST_NAME, a.au_fname FIRST_NAME, sum(((t.ytd_sales*t.royalty/100)*ta.royaltyper/100)s) PROFIT
from 
    authors a
    left join 
        (titleauthor ta   
        inner join 
        (titles t inner join publishers p on t.pub_id = p.pub_id)
        on ta.title_id = t.title_id)
    on a.au_id = ta.au_id

group by AUTHOR_ID
order by PROFIT DESC
limit 3
;
```

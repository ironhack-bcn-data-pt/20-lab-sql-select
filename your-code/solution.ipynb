## Challenge 1 - Who Have Published What At Where?

```python
# What titles each author has published at which publishers?
```

```python
%%sql
select 
    au_id as AUTHOR_ID,
    au_lname as LAST_NAME,
    au_fname as FIRST_NAME,
    title as TITLE,
    pub_name as PUBLISHER
from authors 
    inner join titleauthor 
        using (au_id)
    inner join titles 
        using (title_id)
    inner join publishers 
        using (pub_id)
```

```python
%%sql
select 
    count(*)
from authors 
    inner join titleauthor 
        using (au_id)
    inner join titles 
        using (title_id)
    inner join publishers 
        using (pub_id)
```

## Challenge 2 - Who Have Published How Many At Where?

```python
# How many titles each author has published at each publisher?
```

```python
%%sql
select 
    au_id as AUTHOR_ID,
    au_lname as LAST_NAME,
    au_fname as FIRST_NAME,
    pub_name as PUBLISHER,
    count(title) as TITLE_COUNT
from authors 
    inner join titleauthor 
        using (au_id)
    inner join titles 
        using (title_id)
    inner join publishers 
        using (pub_id)
group by 1,2,3,4
order by TITLE_COUNT desc
```

```python
# To check if your output is correct, sum up the TITLE COUNT column. The sum number should be the same as the total 
# number of records in Table titleauthor
```

```python
%%sql
with title_count as (
    select 
        au_id as AUTHOR_ID,
        au_lname as LAST_NAME,
        au_fname as FIRST_NAME,
        pub_name as PUBLISHER,
        count(*) as TITLE_COUNT
    from authors 
        inner join titleauthor 
            using (au_id)
        inner join titles 
            using (title_id)
        inner join publishers 
            using (pub_id)
    group by 1,2,3,4
    order by TITLE_COUNT desc
)

select 
    sum(TITLE_COUNT) 
from title_count
```

## Challenge 3 - Best Selling Authors

```python
# Who are the top 3 authors who have sold the highest number of titles?
```

```python
%%sql
select 
    au_id as AUTHOR_ID,
    au_lname as LAST_NAME,
    au_fname as FIRST_NAME,
    sum(ytd_sales) as TOTAL
from authors 
    inner join titleauthor 
        using (au_id)
    inner join titles 
        using (title_id)
group by 1,2,3
order by TOTAL desc
limit 3
```

## Challenge 4 - Best Selling Authors Ranking

```python
# Now modify your solution in Challenge 3 so that the output will display all 23 authors instead of the top 3. 
# Note that the authors who have sold 0 titles should also appear in your output (ideally display 0 instead of NULL 
# as the TOTAL). Also order your results based on TOTAL from high to low.
```

```python
%%sql
select 
    au_id as AUTHOR_ID,
    au_lname as LAST_NAME,
    au_fname as FIRST_NAME,
    coalesce(sum(ytd_sales), 0) as TOTAL,
    row_number() over (order by coalesce(sum(ytd_sales), 0) desc) as RANK
from authors 
    left join titleauthor 
        using (au_id)
    left join titles 
        using (title_id)
group by 1,2,3
order by TOTAL desc
```

## Bonus Challenge - Most Profiting Authors

```python
# Authors earn money from their book sales in two ways: advance and royalties. An advance is the money that the 
# publisher pays the author before the book comes out. The royalties the author will receive is typically a percentage
# of the entire book sales. The total profit an author receives by publishing a book is the sum of the advance and the
# royalties.
```

```python
# Given the information above, who are the 3 most profiting authors and how much royalties each of them have received?
```

```python
# If a title has multiple authors, how they split the royalties can be found in the royaltyper column of the 
# titleauthor table. We assume the coauthors will split the advance in the same way as the royalties.
```

```python
%%sql
select * from titleauthor
order by title_id, au_ord
```

```python
%%sql
select * from titles
```

```python
%%sql
select 
    a.au_id as AUTHOR_ID,
    au_lname as LAST_NAME,
    au_fname as FIRST_NAME,
    round(sum(CAST(royaltyper AS REAL)/100 * (coalesce(advance, 0) + CAST(coalesce(royalty, 0) AS REAL)/100 * coalesce(ytd_sales, 0) * coalesce(price, 0))), 2) as PROFIT
from authors a
    inner join titleauthor ta
        using (au_id)
    inner join titles t 
        using (title_id)   
group by 1,2,3
order by PROFIT desc
limit 3
```

### Checking tables names

```python
%%sql
SELECT name FROM sqlite_master
WHERE type='table'
ORDER BY name;
```

### Listing "authors" table

```python
%%sql
select au_id, au_lname, au_fname
from authors
```

### Joining "authors" & "titleauthor"

```python
%%sql
select authors.au_id, authors.au_lname, authors.au_fname
from authors join titleauthor on authors.au_id = titleauthor.au_id
```

### Joining "titles" & "titleauthor"

```python
%%sql
select titles.pub_id, titles.title
from titleauthor join titles on titleauthor.title_id=titles.title_id
```

### Joining previous tables together

```python
%%sql
select authors.au_id, authors.au_lname, authors.au_fname, titles.title 
from authors join titleauthor on authors.au_id = titleauthor.au_id
             join titles on titleauthor.title_id = titles.title_id
```

### Joining the "publishers" one to get the challenge table

```python
%%sql
select authors.au_id, authors.au_lname, authors.au_fname, titles.title, publishers.pub_name
from authors join titleauthor on authors.au_id = titleauthor.au_id
             join titles on titleauthor.title_id = titles.title_id
             join publishers on titles.pub_id = publishers.pub_id
```

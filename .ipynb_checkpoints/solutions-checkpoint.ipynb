## Challenge 1

### Solution 1


```python
%%sql
Select a.au_id as 'AUTHOR ID', 
a.au_lname as 'LAST NAME', a.au_fname as 'FIRST NAME', t.title as 'TITLE', p.pub_name as 'PUBLISHER'
from authors a inner join titleauthor ta on a.au_id = ta.au_id
inner join titles t on ta.title_id = t.title_id
inner join publishers p on t.pub_id = p.pub_id

limit 10
```

```python
%%sql
Select a.au_id as 'AUTHOR ID', 
a.au_lname as 'LAST NAME', a.au_fname as 'FIRST NAME', t.title as 'TITLE', count(t.title) as 'TITLE COUNT' 
where a
from authors a inner join titleauthor ta on a.au_id = ta.au_id
inner join titles t on ta.title_id = t.title_id
inner join publishers p on t.pub_id = p.pub_id

limit 10
```

## Challenge 2


### Solution 2


```python
%%sql
Select a.au_id as 'AUTHOR ID', 
a.au_lname as 'LAST NAME', a.au_fname as 'FIRST NAME', p.pub_name as 'PUBLISHER', COUNT(distinct(t.title_id)) as 'TITLE COUNT'
from authors a inner join titleauthor ta on a.au_id = ta.au_id
inner join titles t on ta.title_id = t.title_id
inner join publishers p on t.pub_id = p.pub_id

group by a.au_id
limit 10
```

## Challenge 3


#### First , I explore the sales table

```python
%%sql
Select *
from sales
limit 5
```

#### And try the sum function, we get the total number of books sold by all authors

```python
%%sql
Select sum(s.qty)
from authors a inner join titleauthor ta on a.au_id = ta.au_id
inner join titles t on ta.title_id = t.title_id
inner join publishers p on t.pub_id = p.pub_id
inner join sales s on s.title_id = t.title_id
```

### Solution (3) : 

```python
%%sql
Select a.au_id as 'AUTHOR ID', 
a.au_lname as 'LAST NAME', a.au_fname as 'FIRST NAME', p.pub_name as 'PUBLISHER', SUM(s.qty) as 'TOTAL SALES'
from authors a inner join titleauthor ta on a.au_id = ta.au_id
inner join titles t on ta.title_id = t.title_id
inner join publishers p on t.pub_id = p.pub_id
inner join sales s on s.title_id = t.title_id

group by a.au_id
order by SUM(s.qty) desc
limit 3
```

## Solución Challenge 4

```python
%%sql
create view Authors_Sl as
select authors.au_id as AUTHOR_ID, au_lname as LAST_NAME,  au_fname as FIRST_NAME , titleauthor.title_id as TITLE_ID
from authors
left join titleauthor
on authors.au_id=titleauthor.au_id
group by AUTHOR_ID
```

```python
%%sql
select AUTHOR_ID,LAST_NAME, FIRST_NAME, ifnull(sum(sales.qty),0) as TOTAL
from Authors_Sl
left join sales
on Authors_Sl.TITLE_ID=sales.title_id
group by AUTHOR_ID
order by TOTAL desc
```
